```python
# We can estimate the coefficient values for our training data using stochastic gradient descent.
# Here it required Learning rate and epochs.
# As you can see, we update each coefficient for each row in the training data,each epoch.
# Coefficient are updated based on the error the model made error=predicted-actual
# b1(t+1)=b1(t)-learning rate*error(t)*x1(t)
# b0(t+1)=b0(t)-learning rate*error(t)
```

```python
# Example forestimating coefficients
# Make a prediction
def predict(row,coefficient):
    yhat=coefficient[0]
    for i in range(len(row)-1):
        yhat+=coefficient[i+1]*row[i]
    return  yhat
```

```python
# Estimate linear regression coefficient using stochastic gradient descent
def coefficient_sto_grad_desc(train,leaning_rate,n_epoch):
    coef=[0.0 for i in range(len(train[0]))]
    print("----------------coefficient--------------")
    print(coef)
    for epoch in range(n_epoch):
        sum_error=0
        for row in train:
            yhat=predict(row,coef)
            error=yhat-row[-1]
            sum_error+=error**2
            coef[0]=coef[0]-learning_rate*error
            for i in range(len(row)-1):
                coef[i+1]=coef[i+1]-learning_rate*error*row[i]
        print('epoch=%d,learning_rate=%.3f , error=%.3f' %(epoch,learning_rate,sum_error))
    return coef
```

```python
# Calculate coefficients
dataset=[[1,1],[2,3],[4,3],[3,2],[5,5]]
learning_rate=0.001
n_epoch=200
coef=coefficient_sto_grad_desc(dataset,learning_rate,n_epoch)
print("coefficients :",coef)
```
